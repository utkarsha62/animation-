```python
top_10 = df['Producers'].str.split(',').explode().value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10.values, y=top_10.index, palette='rainbow')
plt.title('Top producers company  who produce highest no of anime')
plt.xlabel('Count')
plt.ylabel('Producer')
plt.show()
```

```python
# how many anime who only for every one (children , adult , old )


plt.figure(figsize=(14,5))
a = sns.countplot(data=df,x="Rating" , hue = 'Rating')
plt.xlabel("categorization of anime ")
plt.ylabel("No. of reservations")
plt.title("categorization of anime based on ranking ")

for bars in a.containers:
    a.bar_label(bars)
```

54 anime who can watch every age person 

```python
# source of anime and  there anime count

plt .figure(figsize=(20,5))
b = sns.countplot(data = df , x ='Source' , hue = 'Source' )
plt.xlabel('source of anime ')
plt.ylabel('no of anime')
plt.title('source of anime ' , fontsize = 20)

for bars in b.containers:
    b.bar_label(bars)
```

```python
t = df.sort_values(by='English', ascending=False).head(10)
```
